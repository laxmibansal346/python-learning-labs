**Q1. Error:**
Wrong function name (readcsv).

import pandas as pd

df = pd.readcsv('data.csv')

# **Correct Code:**

```
import pandas as pd
df = pd.read_csv('dataset.csv')

df
```

**Explanation:**
Correct function is read_csv() in pandas.

**Q2. DataFrame Creation Issue**

**Error:**
Incorrect dictionary syntax.

data = {'Name': ['Aman','Riya'], 'Age': 20, 21}

df = pd.DataFrame(data)

# correct code

```

import pandas as pd

data = {'Name': ['Aman','Riya'], 'Age': [20, 21]}
df = pd.DataFrame(data)

print(df)
```

**Explanation:**
Each key must have values in list format, but Age was not in list.

**Q3. Indexing Mistake**

**Error:**
Incomplete .loc[] syntax.

df = pd.read_csv('data.csv')

df.loc[1:5,'Salary'

```
#Correct Code
import pandas as pd

df = pd.read_csv('dataset.csv')

df.loc[1:5, 'salary']
```

# **Explanation:**
 Closing bracket ] was missing in .loc[].

### **Q4. Filtering Error**

**Error:**
Incorrect condition syntax.

df[df['Salary' > 50000]]

```
# Correct Code
import pandas as pd

df = pd.read_csv('dataset.csv')

df[df['salary'] > 50000]
```

# **Explanation:**
Condition must be written outside column name with proper brackets.

# Q5. Group-by Syntax Error

 **Error:**
**Incorrect groupby syntax.**

df.groupby['Department'].mean()

```
# Correct Code
import pandas as pd

df = pd.read_csv('dataset.csv')

df.groupby('Department').mean()
```

# Explanation:
groupby uses parentheses ( ) to call the function, not square brackets [ ].
